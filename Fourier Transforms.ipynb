# fourier transform of square wave

```python
import numpy as np
import matplotlib.pyplot as plt
N=1000
y=np.zeros(N)
for i in range(N):
  if(i<500):
    y[i]=1
  else:
    y[i]=-1
#plt.plot(range(N),y)
f=np.fft.rfft(y)
f=abs(f/N)
fr=np.fft.rfftfreq(N)
plt.plot(fr,f)
plt.title("fourier transform of square wave")
plt.xlabel("frequency")
plt.ylabel("fourier transform")
plt.savefig("onea.png")
plt.show()
```

# fourier transform of sawtooth

```python
N=1000
y=np.zeros(N)
for i in range(N):
    y[i]=i
#plt.plot(range(N),y)
f=np.fft.rfft(y)
f=abs(f/N)
fr=np.fft.rfftfreq(N)
plt.plot(fr,f)
plt.title("fourier transform of sawtooth")
plt.xlabel("frequency")
plt.ylabel("fourier transform")
plt.savefig("oneb.png")
plt.show()
```

# fourier transform of modulated sine wave

```python
N=1000
y=np.zeros(N)
for i in range(N):
    y[i]=np.sin(np.pi*i/N)*np.sin(20*np.pi*i/N)  
#plt.plot(range(N),y)
f=np.fft.rfft(y)
f=abs(f/N)
fr=np.fft.rfftfreq(N)
plt.plot(fr,f)
plt.title("fourier transform of modulated sine wave")
plt.xlabel("frequency")
plt.ylabel("fourier transform")
plt.savefig("onec.png")
plt.show()
```
